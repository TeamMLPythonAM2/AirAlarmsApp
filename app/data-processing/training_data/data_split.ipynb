```python
import pandas as pd
import pyarrow as pa
import sklearn.model_selection as ms

DATA_DIR = './'
DATA_BASE_NAME = 'isw_telegram_alarms_weather'
DATA_EXT = '.parquet'
```

```python
data = []
for i in range(1, 4):
    data.append(pd.read_parquet(DATA_BASE_NAME + str(i) + DATA_EXT))
```

```python
full_df = pd.concat(data)
full_df.sort_values(by=['datetime'], inplace=True)
```

```python
full_df.head(50)
```

```python
tss = ms.TimeSeriesSplit(n_splits=2)
train_idx, test_idx = tss.split(full_df)
```

```python
train_df = full_df.iloc[train_idx[1], :]
test_df = full_df.iloc[test_idx[1], :]
```

```python
print(f"Alarms in train df{train_df[train_df['is_alarm'] != 0].shape}\n"
      f"Total size of train df{train_df.shape} \n")
```
