```python
import pandas as pd

final_df = pd.read_parquet('./final_dataset.parquet')
```

```python
isw_columns = [column for column in final_df.columns.to_list() if column.startswith('isw')]
telegram_columns = [column for column in final_df.columns.to_list() if column.startswith('telegram')]
```

```python
final_df.head(550)
```

```python
text_columns = telegram_columns + isw_columns
text_data = final_df[['datetime'] + text_columns]
final_df.drop(columns=text_columns, inplace=True)
```

```python
text_data.loc[:, 'datetime'] += pd.Timedelta(days=1)
text_data = text_data.groupby('datetime').first()
text_data.head(100)
```

```python
final_df = final_df.merge(text_data, on='datetime')
```

```python
final_df.to_parquet('./final_dataset_shifted.parquet')
```
