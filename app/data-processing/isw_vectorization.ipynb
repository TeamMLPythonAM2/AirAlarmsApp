```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import PCA, TruncatedSVD
from app.config.utils import *
from app.config.configuration import Config
import pandas as pd
```

```python
tf_idf = TfidfVectorizer(ngram_range=(2,2), stop_words="english", max_features=10000)
```

```python
df = pd.read_parquet(Config.ISW_PARQUET_PATH)
```

```python
df.head(10)
```

```python
tfidf_matrix = tf_idf.fit_transform(df["text_lemm"])
```

```python
tf_idf.get_feature_names_out()
```

```python
tfidf_matrix.shape
```

```python
pca = PCA(n_components=750)
```

```python
tfidf_matrix_pca = pca.fit_transform(tfidf_matrix)
```

```python
tfidf_matrix_pca.shape
```

```python
tfidf_matrix_pca
```

```python
df["vector"] = tfidf_matrix_pca.tolist()
```
