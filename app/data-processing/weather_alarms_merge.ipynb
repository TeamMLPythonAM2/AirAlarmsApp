```python
import os
import pandas as pd
import datetime as dt
import numpy as np
```

```python
DATA_DIR ='data'
ALARMS_FILE_NAME = 'alarms_clean.csv'
WEATHER_FILE_NAME = 'weather_clean.csv'
REGION_CITY_ADDRESS = 'region_city_address.csv'
ISW_DATA_NAME = 'isw_data.parquet'
```

Alarms

```python
alarms_path = os.path.join(DATA_DIR, ALARMS_FILE_NAME)
```

```python
alarms_df = pd.read_csv(alarms_path)
```

```python
alarms_df["region_city"].unique()
```

```python
alarms_df.drop_duplicates(subset=['region_city', 'start'], inplace=True)
alarms_df.shape
```

Weather

```python
weather_path = os.path.join(DATA_DIR, WEATHER_FILE_NAME)
```

```python
weather_df = pd.read_csv(weather_path)
```

```python
weather_df["city_address"].unique()
```

```python
weather_kyiv_oblast = weather_df.replace({"Kyiv": "Kyiv oblast"}, inplace=False)
weather_kyiv_oblast = weather_kyiv_oblast[weather_kyiv_oblast["city_address"] == "Kyiv oblast"]
```

```python
weather_kyiv_oblast
```

```python
weather_fixed_df = pd.concat([weather_df, weather_kyiv_oblast])
```

```python
weather_fixed_df["city_address"].unique()
```

```python
combined_datetime = pd.to_datetime(weather_fixed_df['day_datetime'], format="%Y-%m-%d") + \
                    pd.to_timedelta(weather_fixed_df['hour_datetime'])
weather_fixed_df['datetime'] = combined_datetime
weather_fixed_df.drop(columns=['day_datetime', 'hour_datetime'], inplace=True)
weather_fixed_df
```

REGION_CITY_ADRESS MERGING TABLE

```python
region_city_address_path = os.path.join(DATA_DIR, REGION_CITY_ADDRESS)
```

```python
region_city_address = pd.read_csv(region_city_address_path)
region_city_address
```

```python
merged_alarms_df = alarms_df.merge(region_city_address, left_on="region_city", right_on="region_city")
```

```python
print(merged_alarms_df.shape, alarms_df.shape)
```

Merging formatted weather and alarms

```python
merged_alarms_df['datetime'] = pd.to_datetime(merged_alarms_df['start']).dt.floor('h')
alarms_and_weather_df = merged_alarms_df.merge(
    weather_fixed_df, 
    how="right",
    left_on=['datetime', 'city_address'], 
    right_on=['datetime', 'city_address']
).fillna(0)
```

```python
exclude_columns = [
    'region_city', 'all_region', 'id', 'region_id', 
    'end', 'start_month', 'start_year', 'start_day',
    'start_hour', 'start_minute', 'duration', 'Unnamed: 0'
]
```

```python
# alarms_and_weather_df.drop_duplicates(subset=["start", "city_address"], inplace=True)
alarms_and_weather_df.drop(columns=exclude_columns, inplace=True)
alarms_and_weather_df['day'] = alarms_and_weather_df['datetime'].apply(lambda date: date.date())
alarms_and_weather_df['is_alarm'] = alarms_and_weather_df['duration_in_min'].map(
    lambda duration: int(duration != 0) 
)
```

```python
alarms_and_weather_df.head(550)
```

```python
alarms_and_weather_df['datetime'] = alarms_and_weather_df['datetime'].apply(lambda date: str(date))
alarms_and_weather_df['start'] = alarms_and_weather_df['start'].apply(lambda date: str(date))
alarms_and_weather_df.to_parquet(os.path.join(DATA_DIR, 'alarms_weather.parquet'))
```

```python
alarms_and_weather_df.columns
```

Getting ISW data

```python
ISW_PATH = os.path.join(DATA_DIR, ISW_DATA_NAME)
isw_data = pd.read_parquet(path=ISW_PATH)
```

```python
# isw_data.drop(columns=['short_url', 'full_url'], inplace=True)
isw_data['date'] = isw_data['date'].apply(lambda date: dt.datetime.strptime(date, '%Y-%m-%d %H:%M:%S').date())
# isw_data.duplicated()
# isw_data.head(5)
```

Merging ISW data

```python
isw_alarms_weather = alarms_and_weather_df.merge(isw_data, how="left", left_on='day', right_on='date')
```

```python
isw_alarms_weather.dropna(subset=isw_data.columns, inplace=True)
isw_alarms_weather.drop(columns=['day', 'date', 'datetime'], inplace=True)
isw_alarms_weather.shape
```

```python
num_parts = 3
chunks = np.array_split(isw_alarms_weather, num_parts)

for i, chunk in enumerate(chunks, start=1):
    output_file = f"isw_alarms_weather{i}.parquet"
    chunk.to_parquet(output_file, compression='snappy')
```
