```python
from sklearn.feature_extraction.text import TfidfVectorizer
from app.config.utils import *
from app.config.configuration import Config
import pandas as pd
```

```python
tf_idf = TfidfVectorizer(ngram_range=(2,2), stop_words="english")
```

```python
df = pd.read_parquet(Config.ISW_PARQUET_PATH)
```

```python
df.head(10)
```

```python
tf_idf.fit(df["text_lemm"])
```

```python
df["vector"] = df["text_lemm"].apply(func=lambda x: tf_idf.transform([x]))
```

```python
df.head()
```

```python
df["vector"] = df["vector"].apply(lambda x: x.toarray().flatten())
```

```python
df.to_parquet(Config.ISW_PARQUET_PATH)
```
